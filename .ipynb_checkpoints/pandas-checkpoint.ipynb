## Pandas Conditional Selection and Data Acessing

```python
import pandas as pd
import numpy as np
```

### Generate Some Random Data

###### How this is done is not important
###### The data set is 100 rows long and contains 2 categorical columns and on numberical column
###### The first 10 rows are displayed below

```python
data_length = 100

data = {
            "trial": np.random.choice(["1", "2", "3", "4"], data_length),
            "cat_col2": np.random.choice(["high", "low"], data_length),
            "num_col1": np.random.randn(data_length) * 100
        }

# the data is stored in a very useful structure called a dataframe (or df)
df = pd.DataFrame(data)

# this displays the first 10 rows of the df
df.head(8)
```

### Conditional Selection

##### The DataFrame.loc method is really useful!

```python
# get return a subset of the dataframe where cat_col1 = A

# only_A = df.loc[df["cat_col1"] == "A"]
# only_A.tail(2)  # display the last two rows

# only_A = df.loc[df["cat_col2"] == "low"][["num_col1", "cat_col2"]]
```

```python
# get only the rows where cat_col2 = "low"

only_low = df.loc[df["cat_col2"] == "low"]
only_low.head(3)
```

#### Multiple Conditions

##### The DataFrame.loc can compine multiple conditions

```python
# multiple conditions:
# get only rows where:
#     - (cat_col1 = A) AND
#     - (cat_col2 = low)

multiple_conditions = df.loc[
                                (df["cat_col1"] == "A") &
                                (df["cat_col2"] == "low")
                            ]
multiple_conditions.head(2)
```

```python
# get only rows where:
#     - (cat_col1 = B) OR
#     - (num_col1 > 50)

mult_cond2 = df.loc[
                        (df["cat_col2"] == "B") |
                        (df["num_col1"] > 50)
                   ]
mult_cond2.tail(2)
```

```python
# get only rows where:
#     - NEITHER
#     - (cat_col1 = B) NOR
#     - (num_col1 > 50)

# this is a bit complex and not that common so dont worry if its not immediately obvious
# the "~" basically acts as in inverse to whatever follows it

mult_cond3 = df.loc[
                        ~(
                        (df["cat_col2"] == "B") |
                        (df["num_col1"] > 50)
                        )
                   ]
mult_cond2.tail(2)
```

### Accessing Data

###### I normally like to keep my data in pandas DataFrames, but if you want to get values from individual columns, heres how you do it

```python
# selecting a single column

only_cat_col1 = df["cat_col1"]
only_cat_col1.head(3)
```

```python
# data from multiple columns

two_cols = df[["cat_col1", "num_col1"]]
two_cols.head(3)
```

#### Acessing values

###### In the above, data was returned to you in pandas data structure, if you really need the actual values (I very rarely do), then you can call the .values attribute

```python
cat_col1_vals = df["cat_col1"].values
cat_col1_vals
```

### Tying it all together

```python
# you can combine multiple steps into one line if you like
# its really not necessary but you might find it useful as you get more comfortable

df.loc[df["num_col1"] > 70]["cat_col1"].head(3)  # same as the above cell
```

```python
def get_trial_statistics(sub_df):
    mean_num_col = np.mean(sub_df["num_col1"])
    std_num_col = np.std(sub_df["num_col1"])
    first_cat_col2 = sub_df["cat_col2"].iloc[0]
    
    return pd.Series(
                        {
                            "mean_num_col": mean_num_col,
                            "std_num_col": std_num_col,
                            "first_cat_col2": first_cat_col2
                        }
                    )
    
```

```python
df.groupby("trial").apply(get_trial_statistics)
```
